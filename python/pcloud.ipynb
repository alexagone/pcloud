```python
import numpy as np
import matplotlib.pyplot as plt

DEFAULT_PCLOUD_CLOUD = "/tmp/pcloud_cloud.csv"
DEFAULT_PCLOUD_HULL = "/tmp/pcloud_hull.csv"
```

Here display some of the relevant internal variables

```python
print(DEFAULT_PCLOUD_CLOUD)
print(DEFAULT_PCLOUD_HULL)
```

```python
cloud = np.genfromtxt(DEFAULT_PCLOUD_CLOUD, delimiter=' ')
hull= np.genfromtxt(DEFAULT_PCLOUD_HULL, delimiter=' ')
```

```python
The following plot display the randomly generated point cloud (in blue) along with the detected hull found by the algorithm (in orange). The hull points have been sorted clockwise and annotated from 0 to N.
```

```python
fig, ax = plt.subplots()
ax.scatter(cloud[:,0], cloud[:,1])
ax.scatter(hull[:,0], hull[:,1])

for i, txt in enumerate(range(len(hull))):
   ax.annotate(txt, (hull[:,0][i], hull[:,1][i]))

plt.show()
```
